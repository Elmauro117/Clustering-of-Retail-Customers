## Catching Outliers

```python
df.columns
```

```python
ncols = 3
nrows = int(np.ceil(len(df.columns) / (1.0*ncols)))
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(25, 25))

counter = 0
for i in range(nrows):
    for j in range(ncols):

        ax = axes[i][j]

        if counter < len(df.columns):

            ax.hist(df[df.columns[counter]], bins=10, color='blue', alpha=0.5, label='{}'.format(df.columns[counter]))
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            #ax.set_ylim([0, 5])
            leg = ax.legend(loc='upper left')
            leg.draw_frame(False)

        # Remove axis when we no longer have data
        else:
            ax.set_axis_off()

        counter += 1

plt.show()
```

#### Distribution looks fine, except for people with 120+ years old & ppl that earn too much money comapred to the rest

```python
## So we proceed to clip that unfathomable outlier
df.loc[df.Age > 100, "Age"] = 88
df.loc[df.Income > 120000, "Income"] = 120000
```

#### Some More Data Engineering

```python
## Hacer un shuffle
from sklearn.utils import shuffle
df=shuffle(df)
```

```python
# dividing age into groups
df['Age'] = pd.cut(x = df['Age'], bins = [18, 44, 64, 100],
                         labels = ['Adult','Senior','SeniorPlus'])
```

```python
## Dividing into more groups and dropping senseless features
df["Education"]=df["Education"].replace({"Basic":"Basic","2n Cycle":"Mid", "Graduation":"Mid", "Master":"Postgraduate", "PhD":"Postgraduate"})
df["Marital_Status"]=df["Marital_Status"].replace({"Married":"Couple", "Together":"Couple", "Widow":"Widow", "Divorced":"No", "Single":"No","Alone":"No"})

df.loc[df.Marital_Status == "YOLO", "Marital_Status"] = np.nan
df.loc[df.Marital_Status == "Absurd", "Marital_Status"] = np.nan

df = df.dropna(subset=["Marital_Status"])
```

```python
df_dummies= df[["Education","Marital_Status","Age"]]

ColEd = pd.get_dummies(df_dummies["Education"])
ColMa = pd.get_dummies(df_dummies["Marital_Status"])
ColAge = pd.get_dummies(df_dummies["Age"])
```

```python
df = df.drop(["Education","Marital_Status","Age"], axis=1)
```

```python
df1 =  df.drop(['Kidhome', 'Teenhome',
       'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
       'Complain', 'Z_CostContact', 'Z_Revenue', 'Response'], axis=1)
dfBi = df[['Kidhome', 'Teenhome',
       'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2',
       'Complain', 'Z_CostContact', 'Z_Revenue', 'Response']]
```

```python
df_mean = df1.mean()
df_std = df1.std()
df_norm = (df1 - df_mean)/df_std
```

```python
dff = pd.concat([df_norm,dfBi,ColEd,ColMa,ColAge],axis = 1)
```

```python
dff['TotalMntSpent'] = dff['MntWines'] + dff['MntFruits'] + dff['MntMeatProducts'] + dff['MntFishProducts'] + dff['MntSweetProducts'] + dff['MntGoldProds']
```

```python
dff['TotalNumPurchases'] = dff['NumWebPurchases'] + dff['NumCatalogPurchases'] + dff['NumStorePurchases'] + dff['NumDealsPurchases']
```

```python
### Esto se peude mirr aparte

dff['Total_Acc_Cmp'] = dff['AcceptedCmp1'] + dff['AcceptedCmp2'] + dff['AcceptedCmp3'] + dff['AcceptedCmp4'] + dff['AcceptedCmp5'] + dff['Response']
```

```python
dff.drop(['MntWines','MntFruits','Z_CostContact','Z_Revenue','MntMeatProducts','MntFishProducts',
          'MntSweetProducts','MntGoldProds','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumDealsPurchases',
          'AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response'], axis=1, inplace=True)
```
